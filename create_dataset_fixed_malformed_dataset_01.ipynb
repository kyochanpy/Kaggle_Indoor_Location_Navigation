```python
from pathlib import Path
import re

d = Path('/kaggle/input/indoor-location-navigation/test')

sites = []

for test_path in d.iterdir():
  with open(test_path) as f:
    for line in f:
      m = re.match(r'.*SiteID:([^\s]+).*', line)
      if m:
        sites.append(m[1])
        break;

sites = set(sites)
print(f'site={len(sites)}')
sites
```
